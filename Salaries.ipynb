```python
!pip install -U gdown

import gdown
file_id = '1PyA-9EpUcNU7hzBuxFOblFe303V-x38H'
gdown.download(f'https://drive.google.com/uc?id={file_id}', 'DataScience_Salaries.csv', quiet=False)

# Cargar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración general de visualización
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Cargar el archivo CSV
df = pd.read_csv('DataScience_Salaries.csv')

# Vista previa de los datos
print("Vista previa del dataset:")
display(df.head())

# Limpieza de datos
df_clean = df.dropna()

# Agrupaciones para análisis
salary_by_exp = df_clean.groupby("experience_level")["salary_in_usd"].mean().sort_values()
salary_by_category = df_clean.groupby("job_category")["salary_in_usd"].mean().sort_values()
salary_by_setting = df_clean.groupby("work_setting")["salary_in_usd"].mean().sort_values()
company_size_counts = df_clean["company_size"].value_counts()

# Gráfico - Salario por experiencia
plt.figure()
sns.barplot(x=salary_by_exp.index, y=salary_by_exp.values)
plt.title("Salario Promedio por Nivel de Experiencia")
plt.ylabel("Salario en USD")
plt.xlabel("Nivel de Experiencia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico - Salario por categoría de puesto
plt.figure()
sns.barplot(x=salary_by_category.values, y=salary_by_category.index)
plt.title("Salario Promedio por Categoría de Puesto")
plt.xlabel("Salario en USD")
plt.ylabel("Categoría")
plt.tight_layout()
plt.show()

# Gráfico - Salario por modalidad de trabajo
plt.figure()
sns.barplot(x=salary_by_setting.index, y=salary_by_setting.values)
plt.title("Salario Promedio por Modalidad de Trabajo")
plt.ylabel("Salario en USD")
plt.xlabel("Modalidad")
plt.tight_layout()
plt.show()

# Gráfico - Distribución por tamaño de empresa
plt.figure()
sns.barplot(x=company_size_counts.index, y=company_size_counts.values)
plt.title("Distribución por Tamaño de Empresa")
plt.ylabel("Cantidad de Registros")
plt.xlabel("Tamaño de Empresa")
plt.tight_layout()
plt.show()


```
