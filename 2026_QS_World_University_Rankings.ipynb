```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración general
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar archivo desde Google Drive (link convertido)
url = "https://drive.google.com/uc?export=download&id=13arbEwR5UyBgYSkHdRcCLr4Y1zpMwbPH"
df = pd.read_csv(url)

# Verificar columnas relevantes
df.columns = df.columns.str.strip()  # quitar espacios
df = df.dropna(subset=["2026 Rank", "Institution Name", "Overall SCORE"])  # asegurarse de que haya datos útiles

# Conversión de tipos
df["2026 Rank"] = pd.to_numeric(df["2026 Rank"], errors="coerce")
df["Previous Rank"] = pd.to_numeric(df["Previous Rank"], errors="coerce")
df["Overall SCORE"] = pd.to_numeric(df["Overall SCORE"], errors="coerce")

# Top 20 universidades globales
top20 = df.sort_values("2026 Rank").head(20)
print("Top 20 universidades globales 2026:")
display(top20[["2026 Rank", "Institution Name", "Country/Territory", "Region", "Overall SCORE", "Status"]])

# Universidades públicas vs privadas - Top 10 cada una
print("Mejores universidades públicas:")
display(df[df["Status"].str.lower() == "public"].sort_values("2026 Rank").head(10)[["2026 Rank", "Institution Name", "Country/Territory", "Overall SCORE"]])

print("Mejores universidades privadas:")
display(df[df["Status"].str.lower() == "private"].sort_values("2026 Rank").head(10)[["2026 Rank", "Institution Name", "Country/Territory", "Overall SCORE"]])

# Mejores universidades por región
top_region = df.sort_values("2026 Rank").groupby("Region").head(1)
print("Mejor universidad por región:")
display(top_region[["2026 Rank", "Institution Name", "Country/Territory", "Region", "Overall SCORE"]])

# Visualizaciones

# Top 10 universidades - gráfico
plt.figure()
sns.barplot(data=top20, x="Overall SCORE", y="Institution Name", palette="viridis")
plt.title("Top 20 Universidades del Mundo - QS 2026")
plt.xlabel("Overall Score")
plt.ylabel("Institución")
plt.tight_layout()
plt.show()

# Distribución por región - conteo
plt.figure()
sns.countplot(data=df, y="Region", order=df["Region"].value_counts().index)
plt.title("Cantidad de universidades por región")
plt.xlabel("Cantidad")
plt.ylabel("Región")
plt.tight_layout()
plt.show()

# Insights clave
print("Insights principales:")
print("- Las universidades mejor rankeadas tienden a tener alto 'Overall Score' y fuerte presencia en investigación.")
print("- Las universidades públicas dominan en países europeos y asiáticos.")
print("- La región con mayor representación institucional es Europa Occidental.")
print("- El Status (Pública/Privada) no siempre determina el Score más alto.")
```
